```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

```python
START_DATE = '2014-06-01'
END_DATE = '2016-06-13'
symbols = ['AAL', 'ALK', 'AVH', 'CEA', 'ZNH', 'VLRS', 'CPA', 'DAL', 'GOL', 'LUV', 'UAL']
all_data = {symbol : pd.read_csv('./csv/{}.csv'.format(symbol), index_col='Date', parse_dates=True) for symbol in symbols}
```

```python
luv_price = all_data['LUV']['Adj Close']
luv_price
```

```python
ma_luv_5 = all_data['LUV'].rolling(window=5).mean()
ma_luv_5
```

```python
ma_luv_25 = all_data['LUV'].rolling(window=25).mean()
ma_luv_25
```

```python
symbols_noluv = symbols.copy()
symbols_noluv.remove('LUV')
avg_daily_return = pd.DataFrame({symbol : all_data[symbol]['Adj Close'] for symbol in symbols_noluv}).pct_change().mean(axis=1)
avg_daily_return.plot(title='daily avg return')
plt.show()
```
