```python
START_DATE = '2014-06-01'
END_DATE = '2016-06-13'
symbols = ['AAL', 'ALK', 'AVH', 'CEA', 'ZNH', 'VLRS', 'CPA', 'DAL', 'GOL', 'LUV', 'UAL']
all_data = {symbol : pd.read_csv('./csv/{}.csv'.format(symbol), index_col='Date', parse_dates=True) for symbol in symbols}
all_data['AAL'].head(2)
```

```python
luvdf = pd.read_csv('LUV.csv', index_col='Date', parse_dates=True)
luvdf.head(2)
```

```python
price = pd.DataFrame({symbol : all_data[symbol]['Adj Close'] for symbol in symbols})
price.head(2)
```

```python
daily_return1 = price.pct_change()
daily_return1
```

```python
s = price.shift(1)
daily_return2 = (price - s) / s
daily_return2
```

```python
aal1 = pd.DataFrame({
    'Return' : all_data['AAL']['Adj Close'],
}).pct_change()
aal1.plot()

aal2 = pd.DataFrame({
    'Volume' : all_data['AAL']['Volume'],
})
aal2.plot()

pd.concat([aal2, aal1]).plot(kind='scatter', x=0, y=0)
plt.show()
```

```python
for symbol in symbols:
    all_data[symbol]['MovingAverage'] = all_data[symbol]['Adj Close'].rolling(window=5).mean()
    
```

```python
onlyluv = all_data['LUV']['Adj Close'].pct_change()
nolue = {symbol : all_data[symbol]['Adj Close'] for symbol in symbols}
nolue.pop('LUV')
nolue = pd.DataFrame(nolue).pct_change().mean(axis=1)
tt = pd.DataFrame({'No.LUV' : nolue, 'LUV' : onlyluv})
tt.plot()
plt.show()
```
